# Scatterplot Map

```python
#importing dependencies
import plotly.graph_objects as go
import pandas as pd
```

```python
#reading csv into dataframe
df = pd.read_csv("../Resources/global_power_plant_database.csv", 
                encoding = "ISO-8859-1", dtype={"fips": str})
```

```python
#creating the scatterplot 
fig = go.Figure(data=go.Scattergeo(
        lon = df['longitude'],
        lat = df['latitude']))

#specifying scope of map
fig.update_layout(
        title = 'Spatial Distribution of Power Plants in the United States',
        geo_scope='usa')
```
