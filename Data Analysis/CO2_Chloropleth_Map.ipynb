# Chloropleth Map CO2 Emissions

```python
#dependencies
import json
import pandas as pd
import plotly.express as px
```

```python
#importing .geojson with geographical coordinates of states
states = json.load(open("../Resources/us-states.json"))
```

```python
#read csv with state CO2 data
df = pd.read_csv("../Output/CO2emissions_by_state.csv", encoding = "ISO-8859-1", dtype={"fips": str})
df = df.drop(columns = ['Unnamed: 0', "Year", "Sector Name"])
#rename State column to code so we can merge with ease
df.rename(columns = {"State":"code"}, inplace = True)
```

```python
#read csv with state abbreviations and long names
abbrev_df = pd.read_csv("../Resources/state_abbreviations.csv", encoding = "ISO-8859-1", dtype={"fips": str})
abbrev_df = abbrev_df.drop(columns = ['abbrev'])
```

```python
#merge with CO2 dataframe
df.merge(abbrev_df, how = "inner", on = "code")
df
```

```python
fig = px.choropleth(df, geojson=states, locations='code', color='CO2 Emissions (Million Metric Tons)')
fig.show()
```
